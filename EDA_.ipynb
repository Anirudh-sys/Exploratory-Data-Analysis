# Data Analysis

```python
# @title 1. Which city has spent the highest amount over the years?
plt.figure(figsize=(12, 5))
city_spend = df.groupby('City')['Amount'].sum().sort_values(ascending=False).head(10)
plt.subplot(1, 2, 1)
colors = plt.cm.Paired(range(len(city_spend)))
city_spend.plot(kind='bar', color=colors, title="Top Cities by Total Spend", ylabel="Total Spend (in millions)", xlabel="City")

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

city_with_highest_spend = city_spend.idxmax()

print(city_with_highest_spend,'has spent the highest amount over the years.')
```

```python
# @title 2. Which card type has the highest amount over the years?

card_spend = df.groupby('Card_Type')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
bubble_sizes = (card_spend.values / card_spend.max()) * 1000
plt.scatter(card_spend.index, card_spend.values, s=bubble_sizes, alpha=0.6, color='green')
plt.title("Total Spend by Card Type (Bubble Plot)")
plt.ylabel("Total Spend (in millions)")
plt.xlabel("Card Type")
plt.grid(True)
plt.show()

highest_card_type = card_spend.idxmax()
print(highest_card_type,'card has the highest amount over the years')

```

```python
# @title 3. Which expense type has the highest amount over the years?

expense_spend = df.groupby('Exp_Type')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
expense_spend.plot(kind='barh', color='orange', title="Total Spend by Expense Type")
plt.xlabel("Total Spend (in millions)")
plt.ylabel("Expense Type")
plt.show()

highest_exp_year = expense_spend.idxmax()
print(highest_exp_year,'expenses have accounted for the highest amounts over the years')
```

```python
# @title 4. What is the total amount spent between males and females (in numbers and percentage)?

gender_spend = df.groupby('Gender')['Amount'].sum()
plt.figure(figsize=(6,6))
plt.pie(gender_spend, labels=gender_spend.index, autopct='%1.1f%%', startangle=90, wedgeprops={'linewidth': 3})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Total Spend by Gender (Doughnut Chart)")
plt.show()

female_spend = gender_spend['F']
male_spend = gender_spend['M']
print(f"Total amount spent by females: {female_spend}","(54.1 %)")
print(f"Total amount spent by males: {male_spend}","(45.9 %)")


```

```python
# @title 5. What is the total amount spent by females via Card Type?

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
female_spend = df[df['Gender'] == 'F'].groupby('Card_Type')['Amount'].sum().sort_values(ascending=False)
female_spend.plot(kind='bar', title="Total Spend by Females by Card Type", ylabel="Total Spend (in millions)", xlabel="Card Type",color = ['lightgreen','violet','grey','orange'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

tot_silver = female_spend['Silver']
tot_gold = female_spend['Gold']
tot_platinum = female_spend['Platinum']
tot_Signature = female_spend['Signature']

print(f"Total amount spent by females via Silver card: {tot_silver}")
print(f"Total amount spent by females via Gold card: {tot_gold}")
print(f"Total amount spent by females via Platinum card: {tot_platinum}")


```

```python
# @title 6. What is the total amount spent by males via Card Type?
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
male_spend = df[df['Gender'] == 'M'].groupby('Card_Type')['Amount'].sum().sort_values(ascending=False)
male_spend.plot(kind='bar', title="Total Spend by Males by Card Type", ylabel="Total Spend (in millions)", xlabel="Card Type",color = ['lightblue','violet','gold','orange'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

tot_silver = male_spend['Silver']
tot_gold = male_spend['Gold']
tot_platinum = male_spend['Platinum']
tot_Signature = male_spend['Signature']

print(f"Total amount spent by Males via Silver card: {tot_silver}")
print(f"Total amount spent by Males via Gold card: {tot_gold}")
print(f"Total amount spent by Males via Platinum card: {tot_platinum}")
```

```python
# @title 7. Top 5 cities with the maximum transactions!

city_transactions = df['City'].value_counts().head(5)
num_colors = len(city_transactions)
colors = sns.color_palette("Blues", num_colors)
city_transactions.plot(kind='bar', title="Top 5 Cities by Number of Transactions", ylabel="Number of Transactions", xlabel="City",color=colors)
plt.xticks(rotation=90)
plt.show()

top_5 = city_transactions
print(top_5)
```

```python
# @title 8. Month-wise spend across the years in descending order

monthly_spend = df.groupby('Month')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
monthly_spend.plot(kind='line', marker='o', color='purple', title="Month-wise Spend in Descending Order (Line Plot)")
plt.ylabel("Total Spend (in millions)")
plt.xlabel("Month")
plt.xticks(ticks=range(len(monthly_spend.index)), labels=monthly_spend.index, rotation=45)
plt.grid(True)
plt.show()

print(monthly_spend)



```

```python
# @title 9. Total amount spent by men via expense type

expense_data = df[df['Gender'] == 'M'].groupby(['Exp_Type', 'Card_Type'])['Amount'].sum().unstack()
expense_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title("Total Amount Spent by Men via Expense Type (Stacked Bar Chart)")
plt.ylabel("Total Amount Spent")
plt.xlabel("Expense Type")
plt.legend(title="Card Type")
plt.show()

print(expense_data)

```
